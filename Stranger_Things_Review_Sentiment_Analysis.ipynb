## Stranger Things Review Sentiment Analysis

## Two good sets of reviews

```python
path = "https://raw.githubusercontent.com/lcchennn/sentiment_analysis/main/stranger_things.csv"
data = pd.read_csv(path, error_bad_lines=False)
data.head()
```

The dataset contains Stranger Things Season 1 and Season 2 reviews form the same group of reviewers.

## Sentiment analysis

```python
from textblob import TextBlob
import nltk
from textblob.sentiments import NaiveBayesAnalyzer
nltk.download('averaged_perceptron_tagger')
from nltk.corpus import stopwords
nltk.download('stopwords')
stops = stopwords.words('english')
```

### Perform sentiment analysis on the first set of reviews.

```python
import nltk
nltk.download('movie_reviews')

reviews = data['Review_1']
index = 1
for review in reviews:
    print(index, "review")
    blob = TextBlob(review,analyzer=NaiveBayesAnalyzer())           #Analyzing Positive and negative sentence 
    for sentence in blob.sentences:
        print(sentence)
        print(sentence.sentiment)
    words_list = [word for word in blob.words if word not in stops] #Deleting stop words
    print(words_list)
    #print(blob.tags)
    #print(blob.noun_phrases)
    
    index+=1
```

## Similarity analysis
### Loading the Language Model and Creating a spaCy Doc


```python
import spacy

# download the spacy model first usnig the following code
# python -m spacy download en_core_web_sm

import en_core_web_sm
nlp = en_core_web_sm.load()
```

## Comparing 2 sets of reviews

### Creating the spaCy Docs

```python
# nlp model needs string data type as argument, so using to_string() to transform series into string

document1 = nlp(data['Review_1'].to_string())
document2 = nlp(data['Review_2'].to_string())
```

```python
simscore = document1.similarity(document2)

print(f'The similarity score between set 1 and set 2 reviews is {simscore:.4f}')
```

## Comparing each review in set 1 with each review in set 2

```python
sim = {}

for i, o in enumerate(data['Review_1']):
    for m, n in enumerate(data['Review_2']):
        similarity = nlp(o).similarity(nlp(n))
        sim[i, m] = round(similarity,4)
#         print(f' Similarity Score between review 1_{i} and review 2_{m} is {similarity}')
```

```python
for k, v in sim.items():
    if k[0] == k[1]:
        print(f'The similarity score between review in set 1 and set 2 by reviewer {k[0]+1} is {v:.4f}')
```
